```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
```

```python
dataset = load_breast_cancer()
```

```python
features_names = dataset.feature_names
print(features_names)
```

```python
target_names = dataset.target_names
print(target_names)
```

1. Malignant - Cancer(0)
2. Benign - Healthy (no cancer)(1)

```python
df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df
```

```python
df.describe()
```

```python
df.head()
```

```python
target = dataset['target']
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size = 0.30, random_state=5)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train,y_train)
from sklearn import metrics
# Get predictions
predicted = model.predict(X_test)
expected = y_test
# Compare results
print (metrics.accuracy_score(expected,predicted))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = np.array(confusion_matrix(y_test , predicted , labels=[1,0]))
confusion = pd.DataFrame(cm,index=['is_cancer','is_healthy'],columns = ['predicted cancer','predicted healthy'])
confusion
```

Recall expresses the ability to find all relevant instances of a class in a data set.

Precision expresses the proportion of the data points our model says existed in the relevant class that were indeed relevant.

```python
print(metrics.classification_report(y_test,predicted))
```

Model correctly classified 90% of the results as malignant. 
Model correctly classified 100% of the results as beningn.

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
model_lr = LogisticRegressionCV(cv=10, max_iter=100000, multi_class='multinomial').fit(X_train, y_train)
```

```python
pred_lr = model_lr.predict(X_test)
pred_lr
```

```python
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
matrix_lr = plot_confusion_matrix(model_lr, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
plt.title('Confusion matrix for LogisticRegression classifier')
plt.show(matrix_lr)
plt.show()
```

```python
print(metrics.classification_report(y_test,pred_lr))
```

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  pred_lr)
auc = metrics.roc_auc_score(y_test, pred_lr)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
